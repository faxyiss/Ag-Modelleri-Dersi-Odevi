```python
import networkx as nx
import matplotlib.pyplot as plt

# Graf oluştur
G = nx.Graph()

# Kenarları ekle (düğüm1, düğüm2, ağırlık)
edges = [
    (1, 2, 4),
    (1, 3, 2),
    (2, 3, 5),
    (2, 4, 10),
    (3, 4, 3),
    (3,5,8),
    (4,5,4),
    (4,6,6),
    (5,6,1)
]

G.add_weighted_edges_from(edges)

# Minumum Yayılan Ağaç hesaplaması
mst = nx.minimum_spanning_tree(G)

# grafın çiziminin belirlenmesi
plt.figure(figsize=(10, 5))
pos = nx.spring_layout(G)  # Düğüm konumlarını belirle

#normal grafın çizdirilmesi
plt.subplot(1, 2, 1)
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=500, font_size=10)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Orijinal Grafik")

#Minumum Yayılan Ağaç Grafının çizdirilmesi
plt.subplot(1, 2, 2)
nx.draw(mst, pos, with_labels=True, node_color="lightgreen", edge_color="orange", node_size=500, font_size=10)
labels_mst = nx.get_edge_attributes(mst, 'weight')
nx.draw_networkx_edge_labels(mst, pos, edge_labels=labels_mst)
plt.title("Minimum Yayılan Ağaç (MST)")

plt.show()
```
